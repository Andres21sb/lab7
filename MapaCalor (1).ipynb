```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Leer el archivo CSV
df = pd.read_csv('/content/School-Age-Digital-Connectivity - Primary.csv')

# Crear una tabla de contingencia (cuenta las frecuencias de combinaciones
#de 'Income Group' y 'Residence-Urban')
contingency_table = pd.crosstab(df['Income Group'], df['Region'])

"""
Crear el mapa de calor usando seaborn:

Como se puede vizualizar, la biblioteca trae la funcion 'heatmap',
la cual nos permite incluso modificar los colores del mapa
dependiendo del resultado que se desea.

annot: Este parametro realiza anotaciones en la celda del mapa,
como podemos vizualizar, en este caso esta seteado en 'True' por
lo cual se pueden vizualizar los numeros en cada cuadricula.

cmap: Permite personalizar el mapa de colores utilizado en la visualización.

"""
#Crear el mapa de calor
plt.figure(figsize=(10, 7))
sns.heatmap(contingency_table, annot=True, cmap="cividis")
# Mostrar el gráfico
plt.title('Mapa de Calor de Income Group vs Residence-Urban')
plt.show()
```

```python
#Tarea de marcadores

CR= folium.Map(location=(10.450691811472405, -84.17604238699441), zoom_start=8)
locations =((9.998551712569991, -84.11111556003829),(9.97108645900616, -84.12840210421459),(10.020955770348, -84.19656278448409),(10.319318906349151, -83.922747075375),
            (8.57351423463531, -82.87239168888978),(9.40275814364291, -83.69151424921571),(10.617696647284287, -85.45140727537121),(10.134755248550517, -85.44678384838853))
popuptexts=('Campus Omar Dengo','Benjamín Núñez', 'Sede interuniversitaria','Campus Sarapiquí',
            'Campus Coto','Campus Pérez Zeledón','Campus Liberia','Campus Nicoya')

#iterar sobre las tuplas para crear los marcadores y los circulos
for location, popuptext in zip(locations,popuptexts):
  popup_html = f'<b>{popuptext}</b>'
  folium.Marker(location=location,popup=popup_html,icon=folium.Icon(color='orange',icon='ok-sign')).add_to(CR)
  folium.Circle(location=location,color='purple',weight=6,fill_color='red',fill_opacity=0.5, tooltip=popuptext).add_to(CR)

CR
```
